```python
'''
pseudoCode Tic-Tac-Toe

Ask Players Name:
Ask begin play(Y/N)

Print X and O - names

[[_,_,_],[_,_,_],[_,_,_]]

player1 = True
player2 = False

play_count=9
while(continue_var)
    
    print the matrix
    
    ask for index
    
    if player1  --> X
    else player2 --> O
    
    1.. index // 4 --> for list num
    2.. index % 3 --> for list element
    
    update list[1][2]
    
    check horizontal
    check vertical
    check diagonal
    
    declare winner
    else continue_var = Wanna continue the game
'''
```

```python
def Play_Tic_Tac_Toe():
    print("Welcome to Tic-Tac_Toe")
    print()

    player1 = input("Player 1 Name:")
    player2 = input("Player 2 Name:")

    print()
    print(f"{player1} is 'X' and {player2} is 'O'")
    print()

    begin_game = input("Would you like to begin? (Y/N)")

    print(begin_game)
    if begin_game != "Y":
        return

    matrix = [["_","_","_"],["_","_","_"],["_","_","_"]]
    player_toggle = True

    for y in range(0,9):
        print(" ".join(matrix[0]))
        print(" ".join(matrix[1]))
        print(" ".join(matrix[2]))

        print()
        print()

        if (player_toggle):
            current_player = player1
            current_input = 'X'
        else:
            current_player = player2
            current_input = 'O'

        print(f"{current_player}, Enter the index (1-9)")
        current_index = int(input())

        input_accepted = True
        while(input_accepted):
            while (not current_index in range(1,10)):
                current_index = int(input("Please enter the correct index between 1-9"))

            list_num = (current_index - 1) // 3
            list_element = (current_index - 1) % 3

            if matrix[list_num][list_element] == "_":
                matrix[list_num][list_element] = current_input
                input_accepted = False
            else:
                current_index = int(input("Already entered, Enter different index"))
                print(f"{current_player}, Enter the index (1-9)")

        result_check = False
        #check horizontal
        for x in matrix:
            if (x[0] == x[1] and x[1] == x[2] and x[2] != "_"):
                result_check = True

        #check Vertical
        for x in range(0,3):
            if (matrix[0][x] == matrix[1][x] and matrix[1][x] == matrix[2][x] and matrix[2][x] != "_"):
                result_check = True

        #check Diagonal
        if (matrix[0][0] == matrix[1][1] and matrix[1][1] == matrix[2][2] and matrix[2][2] != "_"):
            result_check = True

        if (matrix[0][2] == matrix[1][1] and matrix[1][1] == matrix[2][1] and matrix[2][1] != "_"):
            result_check = True

        if y < 9 and not result_check:
            player_toggle = not player_toggle
            next_round = input("Do you wish to continue? (Y/N)")
            if (next_round == 'Y'):
                continue
            else:
                break
        if (result_check):
            print(f"The winner is {current_player}")
            break
    else:
        print("Match Draw!")
```

```python
Play_Tic_Tac_Toe()
```

Player one chooses four random integers. Player two guesses the four integers. In result, show if any integer is correct and if any integer chosen is at correct place

        check if position of the correctly guessed integer is same

```python
from random import randint

list1 = []
list2 = []
guess_value = True
position_value = False
integer_value = False

for x in range(0,4):
    list1.append(randint(0,9))

# print(list1)

while (guess_value):
    
    position_value = False
    integer_value = False

    list2 = [int(x) for x in input("Enter your guess:").split(" ")]
    print(list2)
    
    if (list2 == list1):
        print("Correctly Guessed!!")
        guess_value = False
        break
    
    for (x,y) in enumerate(list2):
        if (y in list1):
            integer_value = True
            
            if (list1[x] == list2[x]):
                position_value = True
                
    if integer_value:
        print("One or more of the Numbers are correct!!")
    if position_value:
        print("One of the integer is in correct place!!")
        
    if (input("Want to continue? Y/N") == 'N'):
        guess_value = False
        print("No Fair! GAME OVER!")
else:
    print(list1)
```
